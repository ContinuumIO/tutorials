least squares

```python
noise = 0.2 * (np.random.random(10) - 0.5)
```

```python
print noise
```

```python
basis = np.random.random((10,3))
```

```python
target = np.dot(basis, [1,2,3])
```

```python
print target
```

```python
noisytarget = target + noise
```

```python
np.linalg.lstsq(basis, noisytarget)
```
